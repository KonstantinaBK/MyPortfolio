```python
#Primero, importaremos pandas y numpy.
```

```python
import pandas as pd
import numpy as np
```

```python
#1. Crea un DataFrame con el archivo entregado.
```

```python
df = pd.read_csv("resumen_resultados_astronautas.csv")
df.head()
```

```python
#2. Crea la columna "IMC" usando la fórmula entregada.IMC = m / h*h
```

```python
df["IMC"] = (df["peso"] / (df["altura"] / 100) ** 2).round(2)
df.head()
```

```python
#3. Crea la columna "promedio_evaluaciones" con el promedio de las 4 columnas
```

```python
df["promedio_evaluaciones"]= df[["evaluacion_fisica", "evaluacion_psicologica", "evaluacion_adaptativa", "evaluacion_medica"]].mean(axis=1)
df.head()
```

```python
#Filtra el DataFrame para obtener solo aquellas filas donde el IMC esté dentro de los límites entregados.

df.query("IMC >= 18.5 and IMC <=24.9")
```

```python
#4. El promedio de las 4 evaluaciones debe ser mayor o igual a 87.
```

```python
df.query("promedio_evaluaciones >= 87")
```

```python
#Filtra el DataFrame resultante y agrega esta condición al filtro aplicado en el paso anterior
```

```python
df= df.query("IMC >= 18.5 and IMC <=24.9 and promedio_evaluaciones >= 87")
df
```

```python
#5. Reinicia el índice del DataFrame resultante
```

```python
df= df.reset_index(drop=True)
df
```

```python
#6. Modifica el valor de la columna "califica", para que diga "Si" en todas las filas del DataFrame resultante. 
```

```python
df["califica"] = "Si"
df
```

```python
#7. Exporta el DataFrame resultante a un archivo .csv separado por comas con el nombre "astronautas_calificados.csv".
```

```python
df.to_csv("astronautas_calificados.csv", index=False, sep=",", header=True, columns=["nombre", "edad", "peso", "altura", "evaluacion_fisica","evaluacion_psicologica", "evaluacion_adaptativa", "evaluacion_medica","anos_experiencia", "califica", "IMC", "promedio_evaluaciones"])
```
