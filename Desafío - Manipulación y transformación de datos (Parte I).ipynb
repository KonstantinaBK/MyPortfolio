```python
#f. Genera una tabla pivote que muestre el número de casos por mes en la fila y por año en las columnas. 
#hint: genere columnas de tipo fecha con pd.to_datetime(). 
#¿Se ve alguna relación entre mes del año y cantidad de casos?


df_merge_all["date"] = pd.to_datetime(df_merge_all["date"])

df_merge_all["Mes"] = df_merge_all["date"].dt.month 

df_merge_all["Año"] = df_merge_all["date"].dt.year

df_pivot_1 = pd.pivot_table(
                                   data = df_merge_all 
                                 , index = "Mes" 
                                 , columns= "Año"
                                 , values = "case_number" 
                                 , aggfunc= "count" 
                                 , margins = True 
)

df_pivot_1
```

```python
#2. Para continuar con el desarrollo de este desafío, necesitarás el archivos Cleaned_DS_Jobs.csv
#a. Carga los datos y crea un DataFrame con ellos.
```

```python
df_cleaned_ds_jobs = pd.read_csv("Cleaned_DS_Jobs.csv")
df_cleaned_ds_jobs
```

```python
#b. Utiliza la siguiente lista de valores que serán considerados como nulos:
#["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"]
```

```python
valores_nulos = ["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"]
valores_nulos
```

```python
#(hint: utiliza el método replace para reemplazar los valores indicados por np.nan)
np.nan, type(np.nan)
df_cleaned_ds_jobs.replace(valores_nulos, np.nan, inplace=True)
df_cleaned_ds_jobs
```

```python
#c. Elimina todas las filas con datos faltantes. (hint: utiliza el método .dropna()
```

```python
df_cleaned_ds_jobs.dropna(axis = 0, how = 'all', inplace=True)
df_cleaned_ds_jobs
```

```python
df_cleaned_ds_jobs.isna().sum(axis=1)
#No se cumple, a pesar de haber utilizado las funciones sugeridas, e incluso haber agregado inplace.
```

```python
#d. A partir de la columna “Salary Estimate”, genera dos columnas: Salario Estimado Mínimo 
#y Máximo. (hint: Utiliza el método apply sobre la columna.)
```

```python
df_cleaned_ds_jobs["Salario Estimado Minimo"] = df_cleaned_ds_jobs["Salary Estimate"].apply(lambda x: int(x.split("-")[0]))
df_cleaned_ds_jobs["Salario Estimado Maximo"] = df_cleaned_ds_jobs["Salary Estimate"].apply(lambda x: int(x.split("-")[1]))
df_cleaned_ds_jobs
```

```python
# 2. e. Realizar la recodificación de la columna Size con los valores de la siguiente tabla
dicc_size = {"10000+ employees": "Mega Empresas",
             "5001 to 10000 employees": "Grandes Empresas",
             "1001 to 5000 employees": "Medianas Empresas",
             "201 to 500 employees": "Pequeñas Empresas",
             "51 to 200 employees": "Pequeñas Grandes Empresas",
             "501 to 1000 employees": "Microempresas",
             "Unknown": "Empresas sin Información"}

df_cleaned_ds_jobs = df_cleaned_ds_jobs.replace(dicc_size)
df_cleaned_ds_jobs

#En el material de estudio, el reemplazo se hace creando un nuevo dataframe. 
```

```python
#f. Finalmente, genera una tabla pivote que muestre la media del salario estimado mínimo y
#la media del salario estimado máximo por tamaño de empresa. (hint: utiliza
#pd.pivot_table para generar la vista adecuada con las columnas generadas.)

df_pivot_3 = pd.pivot_table(
                                   data = df_cleaned_ds_jobs 
                                 , index = "Size"
                                 , values = ["Salario Estimado Minimo", "Salario Estimado Maximo"]
                                 , aggfunc= "mean" 
)

df_pivot_3

#En el diccionario anteriormente reemplazado, no se incluyen las empresas con 1 a 50 trabajadores.
#Queda la duda de por qué no se debe especificar columna, mientras que en las otras tablas, si.
```
