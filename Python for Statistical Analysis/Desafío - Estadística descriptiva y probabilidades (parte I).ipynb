```python
#Importo las librerías. 
import pandas as pd
import numpy as np
```

```python
#Importo el CVS.
df = pd.read_csv("ds_salaries.csv")
df
```

```python
#Reviso las columnas, para trabajar con ellas
df.columns
```

```python
#1. El promedio general de los salarios, desviación estándar, quintiles y rango. 
#(hint: escoge la columna adecuada para comparar)
```

```python
#Promedio general de los salarios (se escogió salarios en dólares, así todos están en una misma moneda).
promedio_gral_salarios= df["salary_in_usd"].mean()
promedio_gral_salarios
```

```python
#Desviación estándar de los salarios.
desviación_estándar= df["salary_in_usd"].std()
desviación_estándar
```

```python
#Quintiles de los salarios.
quintiles=df["salary_in_usd"].quantile([0.20, 0.40, 0.60, 0.80]) 
quintiles
```

```python
#Rango Intercuartil de los salarios.

#Calcular el primer cuartil (Q1) 
q1_salarios = df["salary_in_usd"].quantile(0.25)

#Calcular el tercer cuartil (Q3)
q3_salarios = df["salary_in_usd"].quantile(0.75)
```

```python
#Calcular el rango intercuartil 
iqr_salarios = q3_salarios - q1_salarios
iqr_salarios
```

```python
#2.Una comparación entre los salarios agrupados por 3 diferentes categorías a tu elección. 
#¿Para cuál de ellas las medidas de tendencia central son más representativas? ¿Para cuál son menos representativas? Explica.
```

```python
#Creo un grupo con las columnas por las cuales quiero categorizar el salario.
cat = df.groupby(["experience_level", "employment_type", "remote_ratio"])
```

```python
#Método 1:
```

```python
#Creo un dataframe, para obtener media y mediana del salario, según las categorías. 
#PD: No se puede usar "mode" en función de agregación.
salario_por_categorias = cat["salary_in_usd"].aggregate(["mean","median"])
salario_por_categorias
```

```python
#Moda del salario categorizado
import statistics
moda_salario_por_categorias = cat["salary_in_usd"].agg(statistics.mode).reset_index() 
moda_salario_por_categorias.sort_values(by="salary_in_usd", ascending=False)
```

```python
##Método 2: 
```

```python
mean_salarios_categ = df.groupby(["experience_level", "employment_type", "remote_ratio"])[["salary_in_usd"]].mean()
mean_salarios_categ.sort_values(by="salary_in_usd", ascending=False)
```

```python
median_salarios_categ = df.groupby(["experience_level", "employment_type", "remote_ratio"])[["salary_in_usd"]].median()
median_salarios_categ.sort_values(by="salary_in_usd", ascending=False)
```

```python
mode_salarios_categ = df.groupby(["experience_level", "employment_type", "remote_ratio"])[["salary_in_usd"]].apply(lambda x: x.mode()).reset_index()
mode_salarios_categ.sort_values(by="salary_in_usd", ascending=False)
```

```python
#Método 3, sugerido por compañera. Se deja para servir como material de aprendizaje. 
tabla_pivote1 = pd.pivot_table(df, 
                               values='salary_in_usd', 
                               index='experience_level', 
                               aggfunc=['mean', 'median', 'std','max','min'])
tabla_pivote1
```

```python
tabla_pivote2 = pd.pivot_table(df, 
                               values='salary_in_usd', 
                               index='employment_type', 
                               aggfunc=['mean', 'median', 'std','max','min'])
tabla_pivote2
```

```python
tabla_pivote3 = pd.pivot_table(df, 
                               values='salary_in_usd', 
                               index='remote_ratio', 
                               aggfunc=['mean', 'median', 'std','max','min'])
tabla_pivote3
```

```python
#Para responder, calculamos la dispersión.
std_salarios_categ = df.groupby(["experience_level", "employment_type", "remote_ratio"])[["salary_in_usd"]].std()
std_salarios_categ.sort_values(by="salary_in_usd", ascending=False)
```

```python
#Método 2: Agrupar y filtrar.
best_salary = df.groupby(["company_location", "job_title"])[["salary_in_usd"]].max()
best_salary_US = best_salary.loc["US"] # Filtra solo las filas donde company_location es "US"
best_salary_US.sort_values(by="salary_in_usd", ascending=False)
```

```python
#Método 3: Pivotear y filtrar.
best_salary_by_job = df.pivot_table(index="company_location", columns="job_title", values="salary_in_usd", aggfunc="max")
best_salary_by_job_US = best_salary_by_job.loc["US"] # Filtra solo la fila donde company_location es "US"
best_salary_by_job_US.sort_values(ascending=False)
```
