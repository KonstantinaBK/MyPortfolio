```python
#1. Carga los datos de cada pestaña del archivo US_Regional_Sales_Data.xlsx en un
#DataFrame independiente. (hint: Puedes obtener los nombres de las hojas usando
#pd.ExcelFile(archivo).sheet_names)
```

```python
df_sales = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Sales Orders Sheet')

df_customers = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Customers Sheet')

df_stores = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Store Locations Sheet')

df_products = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Products Sheet')

df_teams = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Sales Team Sheet')
```

```python
#2. Cruza todos los DataFrames usando validación many_to_one, y guarda el resultado en una nueva variable llamada df_base.
```

```python
df_base = df_sales.merge(df_customers, validate = 'many_to_one').\
                   merge(df_stores, validate = 'many_to_one').\
                   merge(df_products, validate = 'many_to_one').\
                   merge(df_teams, validate = 'many_to_one')
```

```python
df_base.head()
```

```python
df_base.info()
```

```python
#3. Agrega las siguientes columnas a df_ base:
#ProcurementDays OrderDate-ProcuredDate
#ShippingDays ShipDate-OrderDate
#DeliveryDays DeliveryDate-ShipDate
#CustomerDays ShippingDays+DeliveryDays
```

```python
df_base["ProcurementDays"]=df_base["OrderDate"]-df_base["ProcuredDate"]
df_base["ShippingDays"]=df_base["ShipDate"]-df_base["OrderDate"]
df_base["DeliveryDays"]=df_base["DeliveryDate"]-df_base["ShipDate"]
df_base["CustomerDays"]=df_base["ShippingDays"]+df_base["DeliveryDays"]
df_base.head()
```

```python
#4. Agrega una nueva columna discreta al df_base, con el nombre CustomerDaysInterval
#que clasifique los valores de la columna CustomerDays en los siguientes intervalos:
#0 to 15 days
#15 to 30 days
#30 to 45 days
#45 to 60 days
#60 to 75 days
#75 to 90 days
```

```python
# Discretizar los datos en intervalos
intervalos = [0, 15, 30, 45, 60, 75, 90]  
# Definir los intervalos deseados
etiquetas = [0, 1, 2, 3, 4, 5]  # Etiquetas numéricas para las categorías

# Convertir la columna CustomerDays en numérica y dividirla por 86400000000000 
df_base["CustomerDays"] = pd.to_numeric(df_base["CustomerDays"]) //86400000000000


# Discretizar y devolver valores en lugar de intervalos
df_base["CustomerDaysInterval"] = pd.cut(df_base["CustomerDays"], bins=intervalos, labels=etiquetas, right=False)

df_base
```

```python
#5. Utiliza una tabla pivote para mostrar el conteo de órdenes agrupando por SalesTeam en las filas y CustomerDaysInterval en las columnas. 
#Este reporte debe cuadrar con los datos de la pestaña Reporte1 del archivo reportes.xlsx
```

```python
df_pivot_1 = pd.pivot_table(
                                   data = df_base 
                                 , index = "Sales Team" 
                                 , columns= "CustomerDaysInterval"
                                 , values = "OrderNumber" 
                                 , aggfunc= "count" 
                                 , margins = True 
)

df_pivot_1
```

```python
#6. Agrega la siguiente columna calculada a la base:
#KPI Cálculo
#GrossMargin OrderQuantity*(UnitPrice*(1-Discount Applied)-UnitCost)
```

```python
df_base.columns
```

```python
df_base["GrossMargin"] = df_base["Order Quantity"]*(df_base["Unit Price"]*(1-df_base["Discount Applied"])-df_base["Unit Cost"])
df_base
```

```python
#7. En esta parte debes aplicar un porcentaje a la columna GrossMargin,
#pero ese porcentaje dependerá del intervalo en que se encuentre GrossMargin.
```

```python
# Discretizar los datos en intervalos
intervalos_margen = [0, 100, 1000, 10000, 100000]  
# Definir los intervalos deseados
etiquetas_margen = [0.05, 0.10, 0.15, 0.20]  # Etiquetas numéricas para las categorías

# Discretizar y devolver valores en lugar de intervalos
df_base["CommissionsPercentage"] = pd.cut(df_base["GrossMargin"], bins=intervalos_margen, labels=etiquetas_margen, right=False)

#Numérico

df_base["CommissionsPercentage"] = pd.to_numeric(df_base["CommissionsPercentage"])

df_base
```

```python
#8. Calcula el monto de la comisión, utilizando la Tabla del Anexo
```

```python
df_base["CommissionsAmount"] = df_base["GrossMargin"]*df_base["CommissionsPercentage"]
df_base
```

```python
#9. Calcula la comisión sobre el margen bruto.
```

```python
df_base["NetMargin"] = df_base["GrossMargin"]-df_base["CommissionsAmount"]
df_base
```

```python
#10. Utilizando una tabla pivote, muestra la suma de los valores GrossMargin, NetMargin y
#CommissionAmount agrupando por Sales Team en las filas. Este reporte debe cuadrar con
#los datos de la pestaña Reporte2 del archivo reportes.xlsx
```

```python
df_pivot_2 = pd.pivot_table(
                                   data = df_base 
                                 , index = "Sales Team"
                                 , values = ["GrossMargin", "NetMargin", "CommissionsAmount"] 
                                 , aggfunc= "sum" 
                                 , margins = True 
)

df_pivot_2
```

```python
#Sugerencia de compañero. Desconozco el motivo por el cual aparece en notación científica, dado que los datos ya son float.
df_pivot_2 = df_pivot_2.applymap(lambda x: "{:.2f}".format(x))
df_pivot_2
```
