```python
#1. Genera una función llamada leer_tabla(tabla, engine) y utilízala para leer tablas
#completas desde la base de datos en dataframes independientes. Utilizando esta función, importa las siguientes tablas:
#- order
#- orderdetails
#- customers
#- products
#- employees
```

```python
# Consulta SQL
query = "SELECT * FROM orders"

# Ejecución de la consulta y carga de resultados en un DataFrame
df_orders = pd.read_sql(query, engine)
df_orders
```

```python
# Consulta SQL
query = "SELECT * FROM orderdetails"

# Ejecución de la consulta y carga de resultados en un DataFrame
df_orderdetails = pd.read_sql(query, engine)
df_orderdetails
```

```python
# Consulta SQL
query = "SELECT * FROM customers"

# Ejecución de la consulta y carga de resultados en un DataFrame
df_customers = pd.read_sql(query, engine)
df_customers
```

```python
# Consulta SQL
query = "SELECT * FROM products"

# Ejecución de la consulta y carga de resultados en un DataFrame
df_products = pd.read_sql(query, engine)
df_products
```

```python
#Otro método

df_employees = pd.read_sql(sql ="employees", con=engine)
df_employees
```

```python
#2. Realiza el cruce entre los DataFrames, asegurándote de utilizar correctamente el parámetro validate para asegurar la integridad referencial.
```

```python
df_base =   df_orders.\
            merge(df_orderdetails, validate='one_to_many').\
            merge(df_products, validate='many_to_one').\
            merge(df_customers, how= "right", validate='many_to_one').\
            merge(df_employees, left_on='salesRepEmployeeNumber', right_on='employeeNumber')

df_base
```

```python
#3. Agrega las siguientes columnas, considerando su nombre y la fórmula asociada
#- venta: quantityOrdered*priceEach
#- costo: quantityOrdered*buyPrice
#- ganancia: considerando las columnas anteriores
```

```python
df_base['venta'] = df_base.quantityOrdered * df_base.priceEach
df_base['costo'] = df_base.quantityOrdered * df_base.buyPrice
df_base['ganancia'] = df_base.venta - df_base.costo
```
